```python
f = codecs.open('LargeTraining.txt', 'r', 'utf-8-sig')
raw= f.read()
```

```python
raw = (raw.replace('*', ' '))
raw = (raw.replace('_', ' '))
raw = (raw.replace('--', ', '))
raw = (raw.replace('[Illustration]', ' '))
raw = (raw.replace("''", '"'))
raw = (raw.replace("``", '"'))
raw = (raw.replace('"', "'"))
```

```python
clean1 = raw.split('\r\n')
```

```python
clean1 = [s for s in clean1 if s != '']
```

```python
space=" "
output = ["{}{}".format(i,space) for i in clean1]
```

```python
CL= open("CleanLarge.txt","w+")
```

```python
CL.writelines(output)
CL.close()
```

```python
## Import the data (one sentence per line)
# path = "/content/gdrive/MyDrive/NLP/HarryPotter.txt"   # harrypotter
# gist = "https://gist.githubusercontent.com/frau-web/ea4a49a223f073d2534850fdef4816fe/raw/9046512cbb2fdd350c8d316ae9d0565621a0b20a/trainingset4.txt"

string = []    
with open("CleanLarge.txt", "r") as CL:
    full_text = CL.read()
    for l in re.split(r"(\.)", full_text):
        if l != ".":
            string.append(l + ". ")

df = pd.DataFrame(string).rename(columns = {0: "text"})
```

```python
df.head()
```

```python
# Clean text
def clean_text(df):
  df = df.drop(df[df["text"].str.isupper()].index) # remove headlines and "THE END" - since written in uppercase letters
  return df    

df = clean_text(df)
```

```python
df.to_csv('Final.txt', index = False, sep='\n')
```

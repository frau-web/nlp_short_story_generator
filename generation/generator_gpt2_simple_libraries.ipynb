## 3. Generating Text Function

```python
def gen_story_gpt2_simple(my_simple_session, my_run, seed, max_len):
  #Change to the correct directory (where the model checkpoint lives)
  %cd "/content/simple/" 
  
  #Generate output
  output = gpt2_simple_library.generate(my_simple_session,
                                    run_name = my_run,
                                    length=max_len,
                                    temperature=0.7,
                                    prefix=seed,
                                    nsamples=1,
                                    batch_size=1,
                                    return_as_list=True
                                    )[0]
  
  #Change back to the default directory of parent notebook
  %cd $code_path 

  #Return output
  return output
```
